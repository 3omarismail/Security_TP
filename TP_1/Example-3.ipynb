```python
import cv2 
import matplotlib.pyplot as plt 
import numpy as np
```

```python
lena = cv2.imread('lena.jpg', cv2.IMREAD_GRAYSCALE)  # Read grayscale
print(lena)
```

```python
keys = [np.random.permutation(256) for _ in range(5)]  # Generate 5 random permutation keys
for i, k in enumerate(keys):
    print(f"Key {i+1} (first 10 values): {k[:10]}")

encrypted_images = []
decrypted_images = []
```

```python
for i, key in enumerate(keys):
    inverse_key = np.argsort(key)           # Inverse key for decryption
    
    encrypted = key[lena]                  # Encrypt
    encrypted_images.append(encrypted)
    
    decrypted = inverse_key[encrypted]     # Decrypt
    decrypted_images.append(decrypted)
```

```python
# Display: 2 rows, 5 columns
plt.figure(figsize=(20, 8))

# Top row: encrypted images
for i, img in enumerate(encrypted_images):
    plt.subplot(2, 5, i+1)
    plt.imshow(img, cmap='gray')
    plt.axis('off')
    plt.title(f'Encrypted {i+1}')

# Bottom row: decrypted images
for i, img in enumerate(decrypted_images):
    plt.subplot(2, 5, i+6)
    plt.imshow(img, cmap='gray')
    plt.axis('off')
    plt.title(f'Decrypted {i+1}')

plt.tight_layout()
plt.show()
```

```python
# Encrypted images look like random noise and hide all visual content.
# Decrypted images perfectly recover the original image.
# Different keys produce different encrypted versions, but decryption always works with the corresponding key.

# Text has a small alphabet and predictable patterns, so simple substitution can be easily cracked.
# Images have 256 possible pixel values and high redundancy, making simple substitution much more effective visually.
```
