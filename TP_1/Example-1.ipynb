```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
lena = cv2.imread('lena.jpg', cv2.IMREAD_GRAYSCALE)  # Read grayscale
print(lena)
```

```python
keys = np.random.randint(1,256,5) # Generate 5 random keys
encrypted_images = []
decrypted_images = []
print(keys)
```

```python
# Encrypt and decrypt
for key in keys:
    enc = (lena + key) % 256       # Encrypt
    dec = (enc - key) % 256        # Decrypt
    encrypted_images.append(enc)   # Store encrypted
    decrypted_images.append(dec)   # Store decrypted
```

```python
fig, axes = plt.subplots(1, len(keys), figsize=(15, 5))
for i, enc in enumerate(encrypted_images):
    axes[i].imshow(enc, cmap='gray')
    axes[i].set_title(f'Encrypted Key {keys[i]}')
    axes[i].axis('off')
plt.tight_layout()
plt.show()
```

```python
# Display decrypted images side by side
fig, axes = plt.subplots(1, len(keys), figsize=(15, 5))
for i, dec in enumerate(decrypted_images):
    axes[i].imshow(dec, cmap='gray')
    axes[i].set_title(f'Decrypted Key {keys[i]}')
    axes[i].axis('off')
plt.tight_layout()
plt.show()
```

```python
# Encrypted images: Each image looks like the original Lena image, but the gray levels are shifted
# Decrypted images: All decrypted images perfectly match the original Lena image
# The Caesar cipher does not change the structure or shapes in the image  you can still recognize the content even after encryption.
```
