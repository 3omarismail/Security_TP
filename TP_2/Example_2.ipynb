```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("lena.jpg", cv2.IMREAD_GRAYSCALE).astype(np.float32)
```

```python
img_dct = cv2.dct(img)
```

```python
img_inverse = cv2.idct(img_dct)
```

```python
# Display results
plt.figure(figsize=(12,4))

plt.subplot(1,3,1)
plt.title("Original")
plt.imshow(img, cmap='gray')
plt.axis("off")

plt.subplot(1,3,2)
plt.title("DCT (log scale)")
plt.imshow(np.log(np.abs(img_dct)+1), cmap='gray')
plt.axis("off")

plt.subplot(1,3,3)
plt.title("Reconstructed (IDCT)")
plt.imshow(img_inverse, cmap='gray')
plt.axis("off")

plt.show()

# Compute difference values
diff = img - img_inverse
print("Max absolute difference:", np.max(np.abs(diff)))
print("Min absolute difference:", np.min(np.abs(diff)))
```

```python
# Compute Mean Square Error (MSE)
mse = np.mean((img - img_inverse) ** 2)

print("MSE =", mse)
```
