```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load gray image
img = cv2.imread("lena.jpg", cv2.IMREAD_GRAYSCALE).astype(np.float32)
```

```python
# 2D-DCT (OpenCV applies 1D-DCT on rows then columns)
img_dct = cv2.dct(img)
```

```python
 #Display
plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
plt.title("Original")
plt.imshow(img, cmap='gray')
plt.axis("off")

plt.subplot(1,2,2)
plt.title("2D-DCT (log scale)")
plt.imshow(np.log(np.abs(img_dct)+1), cmap='gray')
plt.axis("off")

plt.show()
```
