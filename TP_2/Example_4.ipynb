```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load grayscale image
img = cv2.imread("lena.jpg", cv2.IMREAD_GRAYSCALE)
img = img.astype(np.float32)
```

```python
# Compute 2D DCT
dct = cv2.dct(img)

# List of fc values
fc_values = [256, 128, 64, 32, 16, 8, 4, 2]
```

```python
plt.figure(figsize=(12, 8))

for i, fc in enumerate(fc_values):
    # Create a copy for low-pass filtering
    dct_lp = np.zeros_like(dct)
    dct_lp[:fc, :fc] = dct[:fc, :fc]  # keep low frequencies
    
    # Inverse DCT to get filtered image
    img_lp = cv2.idct(dct_lp)
    
    # Display the image
    plt.subplot(2, 4, i+1)
    plt.imshow(img_lp, cmap='gray')
    plt.title(f"fc={fc}")
    plt.axis('off')

plt.suptitle("Low-pass filtering with varying fc")
plt.show()
```
